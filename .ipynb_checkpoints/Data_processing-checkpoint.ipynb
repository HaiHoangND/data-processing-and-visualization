```python
import json
import pandas as pd
from concurrent.futures import ThreadPoolExecutor

# Đọc dữ liệu từ file JSON
with open('Data_homedy.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

# Khởi tạo list chứa dữ liệu cho mỗi cột
# titles = []
# addresses = []
# prices = []
# areas = []
# categories = []
# extensions = []
# project_legals = []
# posting_dates = []
# expire_dates = []
# average_prices_per_m2 = []
# population_densities = []

# Tạo tập hashmap cho population_density
population_density_map = {
    "Ba Đình": 24290, "Bắc Từ Liêm": 7377, "Cầu Giấy": 18800, "Đống Đa": 36284, "Hà Đông": 4149,
    "Hai Bà Trưng": 29647, "Hoàn Kiếm": 27792, "Hoàng Mai": 8346, "Long Biên": 3757, "Nam Từ Liêm": 7234,
    "Tây Hồ": 5442, "Thanh Xuân": 24583, "Sơn Tây": 1604, "Ba Vì": 625, "Chương Mỹ": 1233, "Đan Phượng": 1851,
    "Đông Anh": 1813, "Gia Lâm": 2587, "Hoài Đức": 3250, "Mê Linh": 1357, "Mỹ Đức": 740, "Phú Xuyên": 1092,
    "Phúc Thọ": 1368, "Quốc Oai": 1112, "Sóc Sơn": 978, "Thạch Thất": 884, "Thanh Oai": 1351, "Thanh Trì": 3145,
    "Thường Tín": 1677, "Ứng Hoà": 1054
}

# Hàm xử lý chuỗi giá
def process_price(price_str):
    try:
        # Chuẩn hóa chuỗi
        normalized_str = price_str.replace(',', '.').lower()

        # Tách giá trị và đơn vị
        parts = normalized_str.split('\n')
        value_part = parts[0]
        unit = parts[1]

        # Tính giá trị số nguyên
        if '-' in value_part:
            # Nếu có dấu '-' thì tính trung bình cộng
            start, end = map(float, value_part.split('-'))
            value = (start + end) / 2
        else:
            value = float(value_part)

        # Chuyển đơn vị về giá trị số nguyên
        if unit == 'triệu':
            value *= 10**6
        elif unit == 'tỷ':
            value *= 10**9

        return int(value)
    except Exception as e:
        print("Error:", e)
        return None

# Hàm xử lý chuỗi diện tích
def process_area(area_str):
    try:
        # Chuẩn hóa chuỗi
        normalized_str = area_str.replace(',', '.').lower()

        # Tách giá trị và đơn vị
        parts = normalized_str.split('\n')
        value_part = parts[0]
        unit = parts[1]

        # Tính giá trị số nguyên
        if '-' in value_part:
            # Nếu có dấu '-' thì tính trung bình cộng
            start, end = map(float, value_part.split('-'))
            value = (start + end) / 2
        else:
            value = float(value_part)

        return value
    except Exception as e:
        print("Error:", e)
        return None

# Hàm xử lý category
def process_category(title):
    categories_list = ["căn hộ", "nhà riêng", "nhà mặt phố", "nhà phố thương mại", "biệt thự", "đất nền dự án",
                       "đất", "nhà trọ", "phòng trọ", "trang trại", "khu nghỉ dưỡng", "kho", "nhà xưởng"]
    title_lower = title.lower()
    found_categories = []
    for category in categories_list:
        if category in title_lower:
            found_categories.append(category)
    return found_categories if found_categories else None

# Hàm xử lý extensions
def process_extensions(description):
    extensions_list = ["chiết khấu", "hầm gửi xe", "nhà để xe", "bãi đỗ xe", "hỗ trợ vay", "nội thất",
                       "công viên", "trường học", "bệnh viện", "thang máy", "bể bơi", "khu vui chơi", "gym",
                       "yoga", "cây xanh", "điều hòa", "pccc", "thang thoát hiểm", "đại học", "ban công",
                       "internet", "hỗ trợ lãi suất", "thông minh", "trung tâm thương mại", "chợ",
                       "an ninh", "phòng cháy chữa cháy", "bảo vệ"]
    description_lower = description.lower()
    found_extensions = []
    for extension in extensions_list:
        if extension in description_lower:
            found_extensions.append(extension)
    return found_extensions if found_extensions else None

# Hàm xử lý project_legal
def process_project_legal(description):
    project_legals_list = ["sổ hồng", "sổ đỏ"]
    description_lower = description.lower()
    found_project_legals = []
    for project_legal in project_legals_list:
        if project_legal in description_lower:
            found_project_legals.append(project_legal)
    return found_project_legals if found_project_legals else None

# Duyệt qua từng item trong data để trích xuất thông tin
# for item in data:
#     if item is not None:
#         titles.append(item['title'])
#         addresses.append(item['address'])
#         posting_dates.append(item['posting_date'])
#         expire_dates.append(item['expire_date'])

#         # Xử lý giá
#         price = process_price(item['price'])
#         prices.append(price)

#         # Xử lý diện tích
#         area = process_area(item['area'])
#         areas.append(area)

#         # Xử lý category
#         category = process_category(item['title'])
#         categories.append(category)

#         # Xử lý extensions
#         extension = process_extensions(item['description'])
#         extensions.append(extension)

#         # Xử lý project_legal
#         project_legal = process_project_legal(item['description'])
#         project_legals.append(project_legal)

#         # Tính average_price_per_m2
#         if price is not None and area is not None:
#             average_prices_per_m2.append(price / area)
#         else:
#             average_prices_per_m2.append(None)

#         # Kiểm tra population_density
#         for key, value in population_density_map.items():
#             if key.lower() in item['address'].lower():
#                 population_densities.append(value)
#                 break
#         else:
#             population_densities.append(None)

# # Tạo DataFrame từ dữ liệu
# df = pd.DataFrame({
#     'title': titles,
#     'address': addresses,
#     'price': prices,
#     'area': areas,
#     'category': categories,
#     'extensions': extensions,
#     'project_legal': project_legals,
#     'posting_date': posting_dates,
#     'expire_date': expire_dates,
#     'average_price/1m2': average_prices_per_m2,
#     'population_density': population_densities
# })

# # Lưu DataFrame vào file CSV
# # df.to_csv('Data_homedy_processed.csv', index=False)
# Tạo list chứa dữ liệu cho mỗi item
processed_data = []

# Duyệt qua từng item trong data để trích xuất thông tin và thêm vào processed_data
for item in data:
    if item is not None:
        processed_item = {
            'title': item['title'],
            'address': item['address'],
            'posting_date': item['posting_date'],
            'expire_date': item['expire_date']
        }

        # Xử lý giá
        price = process_price(item['price'])
        processed_item['price'] = price

        # Xử lý diện tích
        area = process_area(item['area'])
        processed_item['area'] = area

        # Xử lý category
        category = process_category(item['title'])
        processed_item['category'] = category

        # Xử lý extensions
        extension = process_extensions(item['description'])
        processed_item['extensions'] = extension

        # Xử lý project_legal
        project_legal = process_project_legal(item['description'])
        processed_item['project_legal'] = project_legal

        # Tính average_price_per_m2
        if price is not None and area is not None:
            processed_item['average_price/1m2'] = round(price / area, 1)
        else:
            processed_item['average_price/1m2'] = None

        # Kiểm tra population_density
        for key, value in population_density_map.items():
            if key.lower() in item['address'].lower():
                processed_item['population_density'] = value
                break
        else:
            processed_item['population_density'] = None

        # Thêm item đã xử lý vào processed_data
        processed_data.append(processed_item)

# Lưu dữ liệu vào file JSON
with open('Data_homedy_processed.json', 'w', encoding='utf-8') as json_file:
    json.dump(processed_data, json_file, ensure_ascii=False, indent=4)
```

```python
with open('Data_dothi.json', 'r', encoding='utf-8') as file:
    data1 = json.load(file)
# Hàm xử lý chuỗi giá
def process_price_dothi(price_str, area):
    try:
        # Chuẩn hóa chuỗi
        normalized_str = price_str.replace(',', '.').lower()
        value_part = ''
        unit = ''
        #Đơn vị có thể là tỷ, triệu/m2, triệu, USD, thỏa thuận
        if 'tỷ' in normalized_str:
            value_part = normalized_str[:-2]
            unit = 'tỷ'
        elif 'triệu/m' in normalized_str:
            value_part = normalized_str[:-8]
            return int(float(value_part)*area*10**6)
        elif 'triệu' in normalized_str:
            value_part = normalized_str[:-5]
            unit = 'triệu'
        # Tính giá trị số nguyên
        if '-' in value_part:
            # Nếu có dấu '-' thì tính trung bình cộng
            start, end = map(float, value_part.split('-'))
            value = (start + end) / 2
        else:
            value = float(value_part)

        # Chuyển đơn vị về giá trị số nguyên
        if unit == 'triệu':
            value *= 10**6
        elif unit == 'tỷ':
            value *= 10**9

        return int(value)
    except Exception as e:
        print("Error:", e)
        return None

# Hàm xử lý chuỗi diện tích
def process_area_dothi(area_str):
    try:
        # Chuẩn hóa chuỗi
        normalized_str = area_str.replace(',', '.').lower()

        # Tách giá trị và đơn vị
        value_part = normalized_str[:-2]

        # Tính giá trị số nguyên
        if '-' in value_part:
            # Nếu có dấu '-' thì tính trung bình cộng
            start, end = map(float, value_part.split('-'))
            value = (start + end) / 2
        else:
            value = float(value_part)

        return value
    except Exception as e:
        print("Error:", e)
        return None
processed_data = []
# Duyệt qua từng item trong data để trích xuất thông tin và thêm vào processed_data
for item in data1:
    if item is not None:
        processed_item = {
            'title': item['title'],
            'address': item['address'],
            'posting_date': item['posting_date'],
            'expire_date': item['expire_date']
        }
        # Xử lý diện tích
        area = process_area_dothi(item['area'])
        processed_item['area'] = area

        # Xử lý giá
        price = process_price_dothi(item['price'], area)
        processed_item['price'] = price

        # Xử lý category
        category = process_category(item['title'])
        processed_item['category'] = category

        # Xử lý extensions
        extension = process_extensions(item['description'])
        processed_item['extensions'] = extension

        # Xử lý project_legal
        project_legal = process_project_legal(item['description'])
        processed_item['project_legal'] = project_legal

        # Tính average_price_per_m2
        if price is not None and area is not None:
            processed_item['average_price/1m2'] = round(price / area, 1)
        else:
            processed_item['average_price/1m2'] = None

        # Kiểm tra population_density
        for key, value in population_density_map.items():
            if key.lower() in item['address'].lower():
                processed_item['population_density'] = value
                break
        else:
            processed_item['population_density'] = None

        # Thêm item đã xử lý vào processed_data
        processed_data.append(processed_item)

# Lưu dữ liệu vào file JSON
with open('Data_dothi_processed.json', 'w', encoding='utf-8') as json_file:
    json.dump(processed_data, json_file, ensure_ascii=False, indent=4)
```
